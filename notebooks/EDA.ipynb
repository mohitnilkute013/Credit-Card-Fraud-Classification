```python
# Select the numerical columns to plot
# numerical_columns = ['Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']
numerical_columns = df.drop(['ID', 'SEX', 'EDUCATION', 'MARRIAGE', 'default payment next month'], axis=1).columns

# Create boxplots for each numerical column
plt.figure(figsize=(15, 15))
for i, column in enumerate(numerical_columns):
    plt.subplot(12, 2, i+1)
    sns.boxplot(data=df[column], orient='h')
    plt.title(f'Boxplot: {column}')
    plt.xlabel(column)
plt.tight_layout()
plt.show()
```
